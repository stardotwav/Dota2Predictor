```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# vari√°vel usada para remover warnings do jupyter notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# leitura e transforma√ß√£o do arquivo em dataframe
dados = pd.read_csv('dadospartidaspro.csv')
dados = dados.drop_duplicates()
dados = dados.drop(columns=['Unnamed: 0', 'id'])
dados
```

```python
# gr√°fico e valor do n√∫mero partidas com vit√≥ria dire e radiant
sns.set_style('darkgrid')
color = sns.color_palette("rocket")
sns.countplot(x='winner', data=dados, palette=color)
```

```python
print(dados[dados['winner']=='radiant']['winner'].count())
print(dados[dados['winner']=='dire']['winner'].count())
```

```python
# matriz de correla√ß√£o entre os dados
sns.heatmap(dados.corr())
```
